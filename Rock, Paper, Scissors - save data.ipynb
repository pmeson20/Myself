```python
import csv 
import random
#Rock: 0     Paper: 1       Scissors: 2
input_data = ['0', '1', '2']

counting = 0

with open('paper.csv', 'a', newline = "") as ofs: 
    
    cwo = csv.writer(ofs) 
    my_list = []
    while counting <=216: 
        
        computer = random.choice(list(input_data)) #Select a random option within the list 
        player = random.choice(list(input_data))
        #player = input("Enter a choice (0, 1 or 2): ") #Ask User for Input 
        
        L = [player, computer]
        my_list.append(L) 
        
        if player == computer: 
            print("Computer's Choice: ", computer)
            print("It's a tie!")

        elif player == '0' and computer == "2": 
            print("Computer's Choice: ", computer)
            print("You have won!") 

        elif player == '1' and computer == "0": 
            print("Computer's Choice: ", computer)
            print("You have won!")

        elif player == "2" and computer == "1": 
            print("Computer's Choice: ", computer)
            print("You have won!") 

        else: 
            print("Computer's Choice: ", computer)
            print("You lost!")

               
        counting = counting + 1
        
    cwo.writerows(my_list) 
```
